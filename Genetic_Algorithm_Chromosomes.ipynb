```python
import numpy as np
class Genetic_Algorithm():
    g=0
    def __init__(self, p1, p2):
        #self.population = np.array(["1000", "0000", "1111","0100"]) 
        self.population = np.array(["1101", "0000", "1111","0100"]) 
        #self.target = "1110"
        self.target = "0110"
        
        print("Initial Population: ", self.population)
        self.crossover_point1 = p1
        self.crossover_point2 = p2
    def binary(self, str):
        number = 0
        j = 3
        for i in range(4):
            if(str[i]=='1'):
                number = number + (pow(2,j))
            j = j - 1
        return number
    
    def fitness_func(self,chromosome):
        f = 0
        for i in range(len(chromosome)):
            if(self.target[i]==chromosome[i]):
                f = f + 1
        return f 
    
    def fitness_chromosomes(self):
        fitness = np.zeros(len(self.population), dtype=int)
        for i in range(len(self.population)):
            fitness[i] = self.fitness_func(self.population[i])   
        print("After Mutation the new population and fitness value is : " )
        for i in range(len(self.population)):
            print("Chromosome-" , i, " : ", self.population[i], "     ", "Fitness Value : " , fitness[i])
      
        
    def uniqueness(self,chromosome):
        flag = 1
        for i in range(len(self.population)):
            if(self.population[i]==chromosome):
                flag =0
                break
        return flag 
    
    def mutation(self, chromosome):
        no_flips = len(chromosome)
        flipped_bit =0 
        c = np.zeros(len(chromosome), dtype=int)
        for i in range(len(chromosome)):
            c[i] = int(chromosome[i])
        while(True):
            if(c[flipped_bit]==0):
                c[flipped_bit] = 1
            else:
                c[flipped_bit] =0
            
            chromosome = ''.join(str(e) for e in c)
            flag = self.uniqueness(chromosome)
            if(flag == 1):
                self.population = np.append(self.population, chromosome)
                break
            flipped_bit = flipped_bit + 1
            no_flips= no_flips-1
            if(no_flips<=0):
                break
        self.fitness_chromosomes()
            
    def crossover(self, chromosome1, chromosome2):
        c1 = np.zeros(len(chromosome1), dtype=int)
        for i in range(len(chromosome1)):
            c1[i] = int(chromosome1[i])
       
        c2 = np.zeros(len(chromosome2), dtype=int)
        for i in range(len(chromosome2)):
            c2[i] = int(chromosome2[i])
       
        j = self.crossover_point2
        
        for i in range(self.crossover_point2-self.crossover_point1+1):
            t = c1[j]
            c1[j] = c2[j]
            c2[j] = t 
            j = j - 1
        
        chromosome1 = ''.join(str(e) for e in c1)
        chromosome2 = ''.join(str(e) for e in c2)
        
        print("After crossover new chromosomes : " ,chromosome1, " , ", chromosome2)
                                   
        flag1=self.uniqueness(chromosome1)
        if(flag1==1):
            # new chromosome
            self.population = np.append(self.population, chromosome1)
            print(self.population)
        else:
            self.mutation(chromosome1)
                                   
        flag2 = self.uniqueness(chromosome2)
        if(flag2==1):
            # new chromosome
            self.population = np.append(self.population, chromosome2)
            print(self.population)
        else:
            self.mutation(chromosome2)
        
    def selection(self):
        print("")
        print("Generation-" , Genetic_Algorithm.g)
        print("=====================")
        sum = 0 
        for i in range(len(self.population)):
            n = self.binary(self.population[i])
            sum = sum + n
        random_n = np.random.randint(0, sum)
        print("Random integer for chromosome1: " , random_n)
        
        # select 1st chromosome
        range1=self.binary(self.population[0])
        pos1 = 0
        for i in range(len(self.population)):
            n = self.binary(self.population[i])
            if(random_n>n and range1<n):
                range1=n
                pos1 = i 
        
        chromosome1 = self.population[pos1]
        self.population = np.delete(self.population, pos1)
        #print(self.population)
        
         # select 2nd chromosome
        sum = 0 
        for i in range(len(self.population)):
            n = self.binary(self.population[i])
            sum = sum + n
        random_n = np.random.randint(0, sum)
        print("Random integer for chromosome2: " , random_n)
        range2=self.binary(self.population[0])
        pos2 = 0
        for i in range(len(self.population)):
            n = self.binary(self.population[i])
            if(random_n>n and range2<n):
                range2=n
                pos2 = i 
       
        chromosome2 = self.population[pos2]
        self.population = np.append(self.population, chromosome1)
        print("Selected chromosomes: ", chromosome1, " , ",chromosome2)
        self.crossover(chromosome1, chromosome2)
        #print(self.population)
        
        flag1=0
        for i in range(len(self.population)):
            if(self.population[i]==self.target):
                flag1 = 1
                break
        if(flag1==0):
            Genetic_Algorithm.g = Genetic_Algorithm.g + 1
            self.selection()
            
        else:
            return 
    
        
print("Enter the crossover point-1: ")
p1 = int(input())
print("Enter the crossover point-2: ")
p2 = int(input())
ob = Genetic_Algorithm(p1, p2)
ob.selection()
```
